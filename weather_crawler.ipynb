```python
import requests
import csv
import time
import pandas as pd
from bs4 import BeautifulSoup

def crawl_weather():
    #craw the weather data from www.hko.gov.hk
    #choose three might useful feature, if raing, max tempreture and min tempreture
    url_prefix = 'https://www.hko.gov.hk/en/wxinfo/pastwx/metob'
    # store two years' data
    weather = pd.DataFrame({},columns=['year','month','day','MaxTemp','MinTemp','isRain'],
              index=range(365*2))

    cnt = 0
    #set url query parameters
    years = ['2017','2018']
    months = ['01','02','03','04','05','06','07','08','09','10','11','12']
    for year in years:
        for month in months:
            url = url_prefix + year + month + '.htm'
            print(url)
            r = requests.get(url)
            html = r.content
            soup = BeautifulSoup(html,'html.parser')
            table = soup.find('table')
            trs = table.find_all('tr')

            for tr in trs[2:-3]:
                weather.iloc[cnt, 0]=year
                weather.iloc[cnt, 1]=month
                tds = tr.find_all('td')

                weather.iloc[cnt, 2]=tds[0].contents[0]
                weather.iloc[cnt, 3]=tds[2].contents[0]
                weather.iloc[cnt, 4]=tds[4].contents[0]
                
                rainLevel = tds[8].contents[0]
                if rainLevel == '-' or rainLevel == 'Trace':
                    weather.iloc[cnt, 5] = 0
                else:
                    weather.iloc[cnt, 5] = 1
                
                cnt += 1
                
    weather.to_csv('weather.csv', index=False)
crawl_weather()
weather = pd.read_csv('weather.csv')
print(weather)
```
